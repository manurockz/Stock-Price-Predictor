```python
data_x=data.copy()
```

```python
data_x.drop(columns=['Close'],axis=0,inplace=True)
```

```python
data_x
```

```python
data_y=data['Close']
```

```python
data_y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(data_x,data_y,train_size=0.8)
```

```python
X_train.shape 
```

```python
y_train.shape
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
model2=linear_model.LinearRegression()
```

```python
model2.fit(X_train,y_train)
```

```python
model2.predict([[208.00,222.25,206.85,216.00,4642146.0,10062.83]])
```

```python
y_test
```

```python
y_predict=pd.Series(model2.predict(X_test))
```

```python
fig2,axes2=plt.subplots(2,figsize=(10,6))
axes2[0].scatter(X_test['Open'].head(100),y_test.head(100))
axes2[1].scatter(X_test['Open'].head(100),y_predict.head(100),color='orange')
```

```python
fig,axes=plt.subplots(2,figsize=(10,6))
axes[0].scatter(X_test['Open'].head(100),y_test.head(100))
axes[1].scatter(X_test['Open'].head(100),y_predict.head(100),color='orange')
axes[0].plot(y_test.head(100),y_predict.head(100),c='orange')
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_predict)
```

```python
import sklearn.metrics as sm
print("Mean absolute error =", round(sm.mean_absolute_error(y_test, y_predict), 2)) 
print("Mean squared error =", round(sm.mean_squared_error(y_test, y_predict), 2)) 
print("Median absolute error =", round(sm.median_absolute_error(y_test, y_predict), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(y_test, y_predict), 2)) 
print("R2 score =", round(sm.r2_score(y_test, y_predict), 2))
```
